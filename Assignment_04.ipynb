```python
import numpy as np
import os
from matplotlib import pyplot as plt

### Tensorflow 2.0 ###
import tensorflow as tf
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras import Input
from tensorflow.keras import layers
from tensorflow.keras import models
from tensorflow.keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras import losses
from tensorflow.keras import optimizers
from tensorflow.keras import metrics
from tensorflow.keras import regularizers
from tensorflow.keras import utils

model = Sequential()
model.add(Input(shape=(128, 128, 1)))
model.add(layers.Conv2D(filters=16, kernel_size=(3, 3), activation = None, padding='same'))
model.add(layers.MaxPool2D())
model.add(layers.Conv2D(filters=32, kernel_size=(3, 3), activation = None, padding='same'))
model.add(layers.MaxPool2D(strides=(2, 2)))
model.add(layers.Conv2D(filters=64, kernel_size=(3, 3), activation = None, padding='same'))
model.add(layers.MaxPool2D(strides=(2, 2)))
model.add(layers.Flatten())
model.add(layers.Dense(128, activation='sigmoid'))
model.add(layers.Dropout(0.25))
model.add(layers.Dense(5, activation='sigmoid'))
tf.keras.utils.plot_model(model, show_shapes=True)
```
